```python
import matplotlib.pyplot as plt
import numpy as np
from lib.reader import Reader
from lib.parameters import Parameters
from itertools import product
```

```python
def generate_parameter_combinations(input_dict):
    """
    Generate all combinations of parameters from the input dictionary.

    Args:
        input_dict (dict): Dictionary containing keys and lists of values.

    Returns:
        list: List of dictionaries representing all combinations of parameters.
    """
    keys = input_dict.keys()
    value_lists = input_dict.values()
    combinations = product(*value_lists)

    result = []
    for values in combinations:
        result.append(dict(zip(keys, values)))

    return result
```
